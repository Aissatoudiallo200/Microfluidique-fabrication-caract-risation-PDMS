```python
import numpy as np
import matplotlib.pyplot as plt
import math


# DONNÉES EXPÉRIMENTALES

P = np.array([15000, 10000, 9000, 7500, 5000, 3000, 2000, 1000], dtype=float)
Q = np.array([1.50e-09, 1.00e-09, 8.33e-10, 7.50e-10,
              5.00e-10, 2.50e-10, 1.67e-10, 6.08e-11], dtype=float)


# RÉGRESSION Q = k * P


k = np.sum(P * Q) / np.sum(P**2)
Rh = 1.0 / k

print(f"Pente k = {k:.3e} (m³/s)/Pa")
print(f"Résistance Rh = {Rh:.3e} Pa·s·m^-3")


# CALCUL DU RAYON EXPÉRIMENTAL


mu0 = 1.0e-3   # viscosité de l'eau (Pa·s)
L = 0.20       # longueur du tube (m)

r_exp = (12 * mu0 * L / (math.pi * Rh))**0.25
D_exp = 2 * r_exp

print(f"Rayon expérimental r = {r_exp:.3e} m")
print(f"Diamètre expérimental D = {D_exp:.3e} m")


# TRACÉ DE Q EN FONCTION DE P


P_line = np.linspace(0, P.max()*1.1, 200)
Q_line = k * P_line

plt.figure()
plt.scatter(P, Q, label="Données expérimentales")
plt.plot(P_line, Q_line, '--', label="Régression linéaire (0,0)")
plt.xlabel("Pression ΔP (Pa)")
plt.ylabel("Débit Q (m³/s)")
plt.title("Q en fonction de ΔP (Tube:resistance de 175 µm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math


# CONSTANTES


mu = 1e-3     # viscosité (Pa·s)
L  = 0.20     # longueur (m)

# 1) RÉSISTANCES THÉORIQUES

# Tube 250 µm
D_250 = 250e-6
r_250 = D_250 / 2
Rh_250_theo = (12 * mu * L) / (math.pi * r_250**4)
r1_4_250_theo = r_250**-0.25

# Tube 175 µm
D_175 = 175e-6
r_175 = D_175 / 2
Rh_175_theo = (12 * mu * L) / (math.pi * r_175**4)
r1_4_175_theo = r_175**-0.25

# Tube 100 µm (théorique)
D_100 = 100e-6
r_100 = D_100 / 2
Rh_100_theo = (12 * mu * L) / (math.pi * r_100**4)
r1_4_100_theo = r_100**-0.25

# 2) RÉSISTANCES EXPÉRIMENTALES


Rh_175_exp = 1.017e13            # Pa·s·m^-3
r_175_exp  = 9.309e-05           # m
r1_4_175_exp = r_175_exp**-0.25

Rh_250_exp = 2.121e12            # Pa·s·m^-3
r_250_exp  = 1.378e-04
r1_4_250_exp = r_250_exp**-0.25


Rh_100_exp = 1.099e14
r_100_exp  = 5.135e-05
r1_4_100_exp = r_100_exp**-0.25


# 3) DROITE THÉORIQUE


x_theo = np.array([r1_4_175_theo, r1_4_250_theo,r1_4_100_theo])
y_theo = np.array([Rh_175_theo, Rh_250_theo,Rh_100_theo]) # Corrected: Removed Rh_100_exp

a_theo, b_theo = np.polyfit(x_theo, y_theo, 1)
x_line_theo = np.linspace(min(x_theo)*0.95, max(x_theo)*1.05, 100)
y_line_theo = a_theo * x_line_theo + b_theo


# 4) DROITE EXPÉRIMENTALE

x_exp = np.array([r1_4_175_exp, r1_4_250_exp,r1_4_100_exp])
y_exp = np.array([Rh_175_exp, Rh_250_exp,Rh_100_exp])

a_exp, b_exp = np.polyfit(x_exp, y_exp, 1)
x_line_exp = np.linspace(min(x_exp)*0.95, max(x_exp)*1.05, 100)
y_line_exp = a_exp * x_line_exp + b_exp


# 5) TRACÉ UNIQUE : théorique + expérimental

plt.figure(figsize=(8,6))


# Droite théorique
plt.plot(x_line_theo, y_line_theo, '-g', label="Droite théorique")

# Points expérimentaux
plt.scatter(x_exp, y_exp, color="red", s=80, label="Points expérimentaux")

# Droite expérimentale
plt.plot(x_line_exp, y_line_exp, '--r', label="Droite expérimentale")



# Axes & affichage
plt.xlabel("r^(1/4) (m^(1/4))")
plt.ylabel("Rh (Pa·s·m^-3)")
plt.title("Droite théorique et expérimentale : Rh en fonction de r^(1/4)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# DONNÉES EXPÉRIMENTALES

P_mbar = np.array([1000, 800, 600, 400, 200])
Q_ul_min = np.array([83.4, 63.8, 45.7, 28.6, 13.5])


# CONVERSIONS

# Convertir pression en Pascal
# 1 mbar = 100 Pa
P = P_mbar * 100

# Convertir Q (µL/min → m³/s)
# 1 µL = 1e-9 m³, 1 min = 60 s
Q = Q_ul_min * 1e-9 / 60


# RÉGRESSION LINÉAIRE FORCÉE À (0,0)

k = np.sum(P * Q) / np.sum(P**2)
Rh = 1 / k

print(f"=== Résultats ===")
print(f"Pente k = {k:.3e} (m³/s)/Pa")
print(f"Résistance fluidique Rh = {Rh:.3e} Pa·s·m⁻³")
print(f"Rh × 4 = {4*Rh:.3e} Pa·s·m^-3")


# TRACÉ Q = f(P)


P_line = np.linspace(0, max(P)*1.1, 200)
Q_line = k * P_line

plt.figure(figsize=(6,4))
plt.scatter(P, Q, label="Mesures")
plt.plot(P_line, Q_line, '--', label="Régression (0,0)")
plt.xlabel("Pression ΔP (Pa)")
plt.ylabel("Débit Q (m³/s)")
plt.title("Débit Q en fonction de la pression ΔP")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# DONNÉES EXPÉRIMENTALES

P_mbar = np.array([1000, 800, 600, 400, 200], float)
Q_ul_min = np.array([49, 37, 26.9, 16, 8], float)


# CONVERSIONS


P = P_mbar * 100                     # Pa
Q = Q_ul_min * 1e-9 / 60             # m³/s


# RÉGRESSION LINÉAIRE FORCÉE PAR (0,0)


k = np.sum(P * Q) / np.sum(P**2)
Rh = 1 / k

print(f"Pente k = {k:.3e} (m³/s)/Pa")
print(f"Résistance fluidique Rh = {Rh:.3e} Pa·s·m^-3")
print(f"Rh × 4 = {4*Rh:.3e} Pa·s·m^-3")


# COURBE Q = f(P)

P_line = np.linspace(0, max(P)*1.1, 200)
Q_line = k * P_line

plt.figure(figsize=(6,4))
plt.scatter(P, Q, label="Mesures expérimentales")
plt.plot(P_line, Q_line, '--', label="Régression linéaire (0,0)")
plt.xlabel("Pression ΔP (Pa)")
plt.ylabel("Débit Q (m³/s)")
plt.title("Débit Q en fonction de la pression ΔP (canal)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math


# CONSTANTES


mu = 1e-3     # viscosité (Pa·s)
L  = 0.20     # longueur (m)


# 1) RÉSISTANCES THÉORIQUES


# Tube 250 µm
D_250 = 250e-6
r_250 = D_250 / 2
Rh_250_theo = (12 * mu * L) / (math.pi * r_250**4)
r1_4_250_theo = r_250**-0.25    # x = r^(-1/4)

# Tube 175 µm
D_175 = 175e-6
r_175 = D_175 / 2
Rh_175_theo = (12 * mu * L) / (math.pi * r_175**4)
r1_4_175_theo = r_175**-0.25

# Tube 100 µm (théorique)
D_100 = 100e-6
r_100 = D_100 / 2
Rh_100_theo = (12 * mu * L) / (math.pi * r_100**4)
r1_4_100_theo = r_100**-0.25


# 2) RÉSISTANCES EXPÉRIMENTALES

Rh_175_exp = 1.017e13            # Pa·s·m^-3
r_175_exp  = 9.309e-05           # m
r1_4_175_exp = r_175_exp**-0.25  # x = r^(-1/4)

Rh_250_exp = 2.121e12            # Pa·s·m^-3
r_250_exp  = 1.378e-04
r1_4_250_exp = r_250_exp**-0.25

Rh_100_exp = 1.099e14            # Pa·s·m^-3
r_100_exp  = 5.135e-05
r1_4_100_exp = r_100_exp**-0.25


# 3) DROITE THÉORIQUE

x_theo = np.array([r1_4_100_theo, r1_4_175_theo, r1_4_250_theo])
y_theo = np.array([Rh_100_theo, Rh_175_theo, Rh_250_theo])

a_theo, b_theo = np.polyfit(x_theo, y_theo, 1)
x_line_theo = np.linspace(min(x_theo)*0.95, max(x_theo)*1.05, 200)
y_line_theo = a_theo * x_line_theo + b_theo


# 4) DROITE EXPÉRIMENTALE (résistances)


x_exp = np.array([r1_4_100_exp, r1_4_175_exp, r1_4_250_exp])
y_exp = np.array([Rh_100_exp, Rh_175_exp, Rh_250_exp])

a_exp, b_exp = np.polyfit(x_exp, y_exp, 1)
x_line_exp = np.linspace(min(x_exp)*0.95, max(x_exp)*1.05, 200)
y_line_exp = a_exp * x_line_exp + b_exp


# 5) PROJECTION DES Rh×4 DES CANAUX SUR LA DROITE EXPÉRIMENTALE


# Rh équivalentes à 20 cm (Rh_canal * 4)
Rh_proj = np.array([5.139e14, 3.000e14, 5.139e14])  # tes trois Rh×4

# Abscisses sur la droite expérimentale : Rh = a_exp * x + b_exp
x_proj = (Rh_proj - b_exp) / a_exp      # x = r^(-1/4)

# Rayon correspondant : r = x^(-4)
r_proj = x_proj**-4

print("=== Points projetés sur la droite expérimentale (par extrapolation) ===")
for i, (Rh_i, x_i, r_i) in enumerate(zip(Rh_proj, x_proj, r_proj), start=1):
    print(f"Canal {i}: Rh20 = {Rh_i:.3e} Pa·s·m^-3, "
          f"x = {x_i:.3e} (r^(-1/4)), "
          f"r = {r_i:.3e} m = {r_i*1e6:.1f} µm")


# 5) TRACÉ UNIQUE : théorique + expérimental + canaux projetés


plt.figure(figsize=(8,6))

# Droite théorique
plt.plot(x_line_theo, y_line_theo, '-g', label="Droite théorique")


# extension automatique de la droite expérimentale 

x_line_exp = np.linspace(
    min(min(x_exp), min(x_proj)) * 0.95,
    max(max(x_exp), max(x_proj)) * 1.05,
    500
)
y_line_exp = a_exp * x_line_exp + b_exp



# Droite expérimentale prolongée
plt.plot(x_line_exp, y_line_exp, '--r', label="Droite expérimentale (prolongée)")

# Points expérimentaux
plt.scatter(x_exp, y_exp, color="red", s=80, label="Résistances expérimentales")

# Points projetés
plt.scatter(x_proj, Rh_proj, color="blue", s=90, label="Canaux (Rh×4 projetés)")
for i in range(len(Rh_proj)):
    plt.text(x_proj[i], Rh_proj[i], f" C{i+1}", color="blue")

plt.ylim(0, 8e14)
plt.xlabel("r^(-1/4) (m^(-1/4))")
plt.ylabel("Rh (Pa·s·m^-3)")
plt.title("Rh en fonction de r^(-1/4) : droites prolongées et canaux")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
